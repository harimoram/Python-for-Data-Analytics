# **About The Dataset**

**Introduction**: In this analysis, we aim to explore a dataset containing information on smartphones, including brands, colors, specifications, and pricing, to gain insights into pricing strategies, consumer behavior, and market trends within the smartphone industry in India.

**Data Description**: The dataset comprises comprehensive information about smartphones, capturing various attributes such as brands, colors, memory, storage, user ratings, selling prices, original prices, and discounts. This dataset provides a rich source of data for analyzing the smartphone market, with detailed specifications and pricing information.

**Loading Sales Dataset**

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('/content/Sales.csv')
```

# **Data Cleaning**

**Handling Missing Values**: Handling missing values depends on various factors such as the nature of the data, the extent of missingness, and the specific requirements of your analysis. Here are some common strategies to handle missing values for the variables mentioned:

```python
# Check for missing values
print(df.isnull().sum())
```

```python
'''Since the number of missing values for both Memory and Storage is relatively low compared to the total number of records,
you can consider imputing these missing values. One simple approach is to impute the missing values with the mean or median of
the respective columns.'''

# Impute missing values in 'Memory' and 'Storage' columns with mean
df['Memory'].fillna(df['Memory'].mean(), inplace=True)
df['Storage'].fillna(df['Storage'].mean(), inplace=True)

'''----OR----'''

'''Droping null values from Memory and Storage'''
# Drop rows with missing values in 'Memory' and 'Storage' columns
df.dropna(subset=['Memory', 'Storage'], inplace=True)
```

```python
'''Rating: If the number of missing values for Rating is substantial, you may want to explore other options for handling them.
One approach is to impute missing values with the mean or median of the Rating column.'''

# Impute missing values in 'Rating' column with median
df['Rating'].fillna(df['Rating'].median(), inplace=True)
```

**Handling Duplicates:** Check for and remove duplicate rows if any.

```python
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

# **Exploratory Data Analysis (EDA)**

**Summary Statistics**: Get an overview of the dataset using summary statistics.

**Correlation Analysis**: Check for correlations between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# **Analysis**

**Visualization-1** Distribution of Ratings Given by Customers

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Ratings Given by Customers')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

**Visualization-2** Distribution of Selling Prices

```python
import matplotlib.pyplot as plt

# Creating a figure for the histogram with specified size
plt.figure(figsize=(8, 6))
# Plotting the histogram of selling prices with 20 bins
plt.hist(df['Selling Price'], bins=20, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.grid(False)
# Displaying the histogram
plt.show()
```

**Visualization-3** Average Discount Percentage by Brand

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Brands', y='discount percentage', data=df, hue='Brands', palette='coolwarm', legend=False)
plt.title('Average Discount Percentage by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=45)
plt.show()
```

**Visualization-4** Sales Performance by Brand

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Brands', y='Selling Price', data=df, estimator=sum, errorbar=None, hue='Brands', palette='muted', legend=False)
plt.title('Sales Performance by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

**Visualization-5** Distribution of Selling Price by Brand- Boxplot

```python
# Boxplot of 'Selling Price' by 'Brands'
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Brands', y='Selling Price', data=df)
plt.title('Distribution of Selling Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Selling Price')
plt.xticks(rotation=45)
plt.show()
```
